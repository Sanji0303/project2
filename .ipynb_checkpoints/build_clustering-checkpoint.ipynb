```python
# ================= IMPORT =================
import pandas as pd
import joblib

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# ================= LOAD DATA =================
df = pd.read_csv("cleaned_data_2.csv")

# ================= FEATURE =================
features = [
    "gia_ban_num",
    "dien_tich_num",
    "price_per_m2",
    "quan_encoded"
]

X = df[features]

# ================= SCALING =================
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ================= KMEANS =================
best_k = 6

kmeans = KMeans(
    n_clusters=best_k,
    random_state=42,
    n_init=10  # tránh warning sklearn
)

kmeans.fit(X_scaled)

# ================= ADD CLUSTER =================
df["cluster_kmeans"] = kmeans.predict(X_scaled)

# ================= SAVE MODEL =================
joblib.dump(kmeans, "kmeans_model.pkl")
joblib.dump(scaler, "scaler_kmeans.pkl")
joblib.dump(features, "features_kmeans.pkl")

print("✅ Saved KMeans model, scaler, features!")
```
