```python
# ================= IMPORT =================
import pandas as pd
import joblib

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# ================= LOAD DATA =================
df = pd.read_csv("cleaned_data_2.csv")

# ================= FEATURE =================
features = [
    "gia_ban_num",
    "dien_tich_num",
    "price_per_m2",
    "quan_encoded"
]

X = df[features]

# ================= SCALING =================
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ================= KMEANS =================
best_k = 6

kmeans = KMeans(
    n_clusters=best_k,
    random_state=42,
    n_init=10  # tránh warning sklearn
)

kmeans.fit(X_scaled)

# ================= ADD CLUSTER =================
df["cluster_kmeans"] = kmeans.predict(X_scaled)

# ================= SAVE MODEL =================
joblib.dump(kmeans, "kmeans_model.pkl")
joblib.dump(scaler, "scaler_kmeans.pkl")
joblib.dump(features, "features_kmeans.pkl")

print("✅ Saved KMeans model, scaler, features!")
```

```python
# export_models.py
import pandas as pd
import joblib
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# ================= LOAD DATA =================
df = pd.read_csv("cleaned_data_2.csv")  # Đảm bảo đường dẫn đúng

# ================= DEFINE FEATURES =================
features = [
    "gia_ban_num",
    "dien_tich_num", 
    "price_per_m2",
    "quan_encoded"
]

# ================= PREPARE DATA =================
X = df[features]

# ================= SCALING =================
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ================= KMEANS =================
best_k = 6
kmeans = KMeans(
    n_clusters=best_k,
    random_state=42,
    n_init=10
)
kmeans.fit(X_scaled)

# ================= ADD CLUSTER TO DATAFRAME =================
df["cluster_kmeans"] = kmeans.predict(X_scaled)

# ================= SAVE MODELS =================
joblib.dump(kmeans, "kmeans_model.pkl")
joblib.dump(scaler, "scaler_kmeans.pkl")
joblib.dump(features, "features_kmeans.pkl")

# ================= VERIFY =================
print("="*50)
print("✅ ĐÃ XUẤT 3 FILE THÀNH CÔNG!")
print("="*50)
print("\n📁 Danh sách file đã xuất:")
print("1. kmeans_model.pkl")
print("2. scaler_kmeans.pkl") 
print("3. features_kmeans.pkl")

# Kiểm tra dung lượng
import os
for file in ["kmeans_model.pkl", "scaler_kmeans.pkl", "features_kmeans.pkl"]:
    if os.path.exists(file):
        size_mb = os.path.getsize(file) / (1024**2)
        print(f"\n📊 {file}: {size_mb:.2f} MB")
```
