```python
# check_clusters.py
import pandas as pd
import joblib

# Load data
df = pd.read_csv("cleaned_data_2.csv")

# Load models
kmeans = joblib.load("file_pkl_bt2/kmeans_model.pkl")
scaler = joblib.load("file_pkl_bt2/scaler_kmeans.pkl")
features = joblib.load("file_pkl_bt2/features_kmeans.pkl")

# Chuẩn bị dữ liệu
X = df[features]
X_scaled = scaler.transform(X)

# Dự đoán cluster
df['cluster'] = kmeans.predict(X_scaled)

# Thống kê từng cụm
print("="*60)
print("THỐNG KÊ CHI TIẾT CÁC CỤM")
print("="*60)

for cluster in sorted(df['cluster'].unique()):
    cluster_data = df[df['cluster'] == cluster]
    print(f"\n📊 CỤM {cluster}:")
    print(f"   Số lượng: {len(cluster_data)} BĐS")
    print(f"   Giá trung bình: {cluster_data['gia_ban_num'].mean()/1e9:.1f} tỷ")
    print(f"   Giá min - max: {cluster_data['gia_ban_num'].min()/1e9:.1f} - {cluster_data['gia_ban_num'].max()/1e9:.1f} tỷ")
    print(f"   Diện tích TB: {cluster_data['dien_tich_num'].mean():.0f} m²")
    print(f"   Diện tích min - max: {cluster_data['dien_tich_num'].min():.0f} - {cluster_data['dien_tich_num'].max():.0f} m²")
    
    # Thống kê theo quận
    quan_stats = cluster_data['quan'].value_counts()
    print(f"   Phân bố theo quận: {dict(quan_stats)}")
    
    # Giá theo m2
    print(f"   Giá/m2 TB: {cluster_data['price_per_m2'].mean():.0f} triệu/m²")

print("\n" + "="*60)
print(f"Tổng số cụm: {len(df['cluster'].unique())}")
```
